# Testing

In order to test the quality of domain shifting of medical images, we are using a common classification task for embryo images. We traied an CNN classifier that perfom almost perfecr (with AUC of 0.989) on the target distribution dataset ED4 and then test thia clssifier perfomance on the shifted dustribustion input dataset (ED1 shifted to the distribution of ED4).

The better the CNN classifier on ED1 shifted, the better our cycle GAN based destribution shift surved it's purpose.

For a more comprehensive test. I would add more typical classification and detection tasks on the shifted dataset (when training is on ED4).
